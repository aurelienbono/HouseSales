

## Data

The dataset contains records of previous houses sold in the area.

| Column Name | Criteria                                                |
|-------------|---------------------------------------------------------|
| house_id    | Nominal. </br> Unique identifier for houses. </br>Missing values not possible. |
| city        | Nominal. </br>The city in which the house is located. One of 'Silvertown', 'Riverford', 'Teasdale' and 'Poppleton'. </br>Replace missing values with "Unknown". |
| sale_price  | Discrete. </br>The sale price of the house in whole dollars. Values can be any positive number greater than or equal to zero.</br>Remove missing entries. |
| sale_date   | Discrete. </br>The date of the last sale of the house. </br>Replace missing values with 2023-01-01. |
| months_listed  | Continuous. </br>The number of months the house was listed on the market prior to its last sale, rounded to one decimal place. </br>Replace missing values with mean number of months listed, to one decimal place. |
| bedrooms    | Discrete. </br>The number of bedrooms in the house. Any positive values greater than or equal to zero. </br>Replace missing values with the mean number of bedrooms, rounded to the nearest integer. |
| house_type   | Ordinal. </br>One of "Terraced" (two shared walls), "Semi-detached" (one shared wall), or "Detached" (no shared walls). </br>Replace missing values with the most common house type. |
| area      | Continuous. </br>The area of the house in square meters, rounded to one decimal place. </br>Replace missing values with the mean, to one decimal place. |


```python

import pandas as pd 
import numpy as np
from IPython.display import Image
```

```python
df = pd.read_csv('data/house_sales.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
missing_values = df.isnull().sum()
missing_values
```

```python
df["city"].fillna("Unknown", inplace=True)
df["sale_date"].fillna("2023-01-01", inplace=True)
df["months_listed"].fillna(df["months_listed"].mean(), inplace=True)
df["bedrooms"].fillna(df["bedrooms"].mean(), inplace=True)
df["house_type"].fillna(df["house_type"].mode(), inplace=True)

df['area'] = df['area'].apply(lambda x : x.split(' ')[0]).astype(float)
df["area"].fillna(df["area"].mean(), inplace=True)
```

```python

missing_values = df.isnull()
print(missing_values.sum())
```

```python
df['house_type'].value_counts()
```

```python
df['house_type'] = df['house_type'].replace(['Det.', 'Semi', 'Terr.'], ['Detached', 'Semi-detached', 'Terraced'])
```

```python
df['house_type'].value_counts()
```

```python
df.info()
```

```python
grouped_df = df.groupby('bedrooms')


price_by_rooms = pd.DataFrame()
price_by_rooms['avg_price'] = grouped_df['sale_price'].mean().round(1)
price_by_rooms['var_price'] = grouped_df['sale_price'].var().round(1)


price_by_rooms
```

```python
# Use this cell to write your code for Task 4

from sklearn.preprocessing import LabelEncoder


from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error



from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python

def wrangle(path_df): 
    df_train =  pd.read_csv(path_df)
    df_train['city'] = LabelEncoder().fit_transform(df_train['city'])
    df_train['house_type'] = LabelEncoder().fit_transform(df_train['house_type'])
    
    df_train['sale_date_year'] = df_train['sale_date'].apply(lambda x : x.split('-')[0]).astype(int)
    df_train['sale_date_month'] = df_train['sale_date'].apply(lambda x : x.split('-')[1]).astype(int)
    df_train['sale_date_day'] = df_train['sale_date'].apply(lambda x : x.split('-')[2]).astype(int)
    
    df_train.drop(columns=['sale_date'],inplace=True)

    
    return df_train 

```

```python
train_data = wrangle('data/train.csv')
train_data.drop(columns=['house_id'],inplace=True)
train_data.head()
```

```python
validation_data = wrangle('data/validation.csv')
validation_data_house_id = validation_data['house_id']
validation_data.drop(columns=['house_id'],inplace=True)
validation_data
```
