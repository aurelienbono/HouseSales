```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("data/insurance.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.describe()
```

```python
plt.figure(figsize=(5,5))
style.use('ggplot')
sns.countplot(x='sex', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x='smoker', data=df)
plt.title('Smoker')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x='region', data=df)
plt.title('Region')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x='region', y='charges', data=df)
plt.title('Cost vs Region')
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x='sex', y='charges',hue='smoker', data=df)
plt.title('Charges for smokers')
```

```python
fig, axes = plt.subplots(1,3, figsize=(15,5), sharey=True)
fig.suptitle('Visualizing categorical columns')
sns.boxenplot(x='smoker', y= 'charges', data=df, ax=axes[0])
sns.boxenplot(x='sex', y= 'charges', data=df, ax=axes[1])
sns.boxenplot(x='region', y= 'charges', data=df, ax=axes[2])
```

```python
df[['age','bmi','children','charges']].hist(bins=30, figsize=(10,10), color='blue')
plt.show()
```

```python
df.head()
```

```python
df['sex'] = df['sex'].apply({'male':0, 'female':1}.get)
df['smoker'] = df['smoker'].apply({'yes':1, 'no':0}.get)
df['region'] = df['region'].apply({'southwest':1, 'southeast':2, 'northwest':3, 'northeast':4}.get)
```

```python
df.head()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(), annot = True)
plt.show()
```

```python
Image('images/yes.jpeg')
```

```python
X = df.drop(['charges', 'sex'], axis=1)
y = df.charges
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
print("X_train shape: ", X_train.shape)
print("X_test shape: ", X_test.shape)
print("y_train shpae: ", y_train.shape)
print("y_test shape: ", y_test.shape)
```

```python
Image('images/train_split.png')
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(X_train, y_train)
pred = linreg.predict(X_test)
```

## EVALUATION MODEL 

```python
Image('images/rr.png')
```

```python
from sklearn.metrics import r2_score
```

```python
print("R2 score: ",(r2_score(y_test, pred)))
```

```python
plt.scatter(y_test, pred)
plt.xlabel('Y test')
plt.ylabel('Y pred')
plt.show()
```
